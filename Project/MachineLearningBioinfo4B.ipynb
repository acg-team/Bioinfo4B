## Prepare for the K means Clustering

We'll save the mutations and the type of the cancer seperately. First the mutations:

```python
mutations = data[["SV", "MSI"]]
```

Did it work?

Now we'll save the type under the variable name label.

## Clustering Time!!!

Below is the basic code. Let's work on, what each parameter means: 

```python
kmeans = KMeans(n_clusters=2, init='random', max_iter=10, n_init=1)
prediction = kmeans.fit_predict(mutations)
plt.scatter(data["SV"], data["MSI"], c = prediction, cmap = "rainbow")
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='black')
plt.xlabel("SV") 
plt.ylabel("MSI")
plt.show() 
```

Finally, we'll check if we labelled the data points correctly:

```python
correct_labels = sum(label == kmeans.labels_)
print("Result: %d out of %d samples were correctly labeled." % (correct_labels, label.size ))
```
